Disciplina: C√°lculo Num√©rico Enunciado Voc√™ dever√° implementar, em Python, o m√©todo de Jacobi para resolver sistemas lineares do tipo ùê¥ùë•=ùëè.

Parte 1 ‚Äì Fun√ß√£o do m√©todo Crie uma fun√ß√£o em Python que:

receba a matriz ùê¥ , o vetor ùëè , um chute inicial, a toler√¢ncia e o n√∫mero m√°ximo de itera√ß√µes;

calcule as itera√ß√µes do m√©todo de Jacobi;

retorne:

a solu√ß√£o aproximada,o n√∫mero de itera√ß√µes realizadas,e mostre na tela, passo a passo, cada itera√ß√£o (valores de ùë• e erro).

Parte 2 ‚Äì Teste do algoritmo Na parte principal do programa (fora da fun√ß√£o):

Defina pelo menos dois sistemas lineares diferentes (matrizes ùê¥ e vetores ùëè );

Para cada sistema, chame a fun√ß√£o criada e exiba o resultado final.

Requisitos O c√≥digo deve estar organizado com a fun√ß√£o em um bloco e as chamadas da fun√ß√£o em outro.

Em cada itera√ß√£o, mostrar algo como:

Itera√ß√£o 1: x = [ ... ] erro = ...

Itera√ß√£o 2: x = [ ... ] erro = ...

ou uma tabela com os resultados de cada intera√ß√£o

Objetivo da atividade Compreender o funcionamento do m√©todo iterativo de Jacobi, implementando o algoritmo e analisando sua converg√™ncia atrav√©s das aproxima√ß√µes sucessivas.

```python
import numpy as np  # biblioteca usada para trabalhar com contas matriciais

# PARTE 1 ‚Äî Procedimento do m√©todo de Jacobi

def metodo_jacobi(M, v, chute, limite, repet):
    # transformar tudo em float para evitar erros num√©ricos
    M = M.astype(float)
    v = v.astype(float)
    atual = chute.astype(float)
    tam = len(v)  # quantidade de inc√≥gnitas

    registro = []  # vai guardar as informa√ß√µes de cada passo

    # salva o estado inicial (itera√ß√£o zero)
    erro_inicial = np.max(np.abs(atual))
    registro.append((0, atual.copy(), erro_inicial))

    # la√ßo principal do processo
    for passo in range(1, repet + 1):
        prox = np.zeros_like(atual)  # recebe os novos valores calculados

        # percorre uma equa√ß√£o por vez
        for linha in range(tam):
            termo = sum(M[linha, col] * atual[col] for col in range(tam) if col != linha)
            prox[linha] = (v[linha] - termo) / M[linha, linha]

        erro_atual = np.max(np.abs(prox - atual))  # maior diferen√ßa entre uma itera√ß√£o e outra

        registro.append((passo, prox.copy(), erro_atual))

        # crit√©rio de parada
        if erro_atual < limite:
            return prox, passo, registro

        # atualiza o vetor para a pr√≥xima rodada
        atual = prox

    # se esgotou o n√∫mero de repeti√ß√µes, devolve o √∫ltimo estado mesmo assim
    return atual, repet, registro
```

```python
import numpy as np

# PARTE 2 ‚Äî Exibi√ß√£o organizada dos resultados
# Cabe√ßalhos e formata√ß√£o da tabela final
def mostrar_resultados(mat, vet, resposta, etapas):
    print("Processo de Jacobi")
    print("Matriz Estendida")

    # imprime cada linha da matriz acompanhada do vetor independente
    for i in range(len(vet)):
        linha_formatada = " ".join(f"{mat[i,j]:7.4f}" for j in range(len(vet)))
        print(f"{linha_formatada} {vet[i]:7.4f}")

    print("\n it    v1       v2       v3       erro")  # cabe√ßalho da tabela

    for it, valores, dif in etapas:
        print(f"{it:2d}  {valores[0]:7.4f}  {valores[1]:7.4f}  {valores[2]:7.4f}  {dif:7.4f}")

    print("\nResultado Final")
    for item in resposta:
        print(f"   {item:.4f}", end="")
    print("\n")
```

```python
#PARTE 3 - Teste
# Sistema para teste

#Matriz A do sistema
A = np.array([
    [3, 1, 1],
    [1, -2, 2],
    [1, -1, 3]
], dtype=float)

b = np.array([8, 3, 8], dtype=float) #Vetor b.

x0 = np.zeros(3) #Chute inicial x0

#Chama o m√©todo com crit√©rio de parada ou toler√¢ncia de 10^(-4).
sol, it, tabela = metodo_jacobi(A, b, x0, limite=1e-4, repet=50)

mostrar_resultados(A, b, sol, tabela)
```
